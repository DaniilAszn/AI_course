## Задание 4.3.1

Задайте массив случайных значений из интервала(0; 1). Рассчитайте средние и медианные значения для массива, сравните результаты, какие выводы можно сделать о значениях? Постройте точечную диаграмму рассения полученного ряда

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.random.sample(10)
index = list(range(len(x)))

fig = plt.figure(figsize = (7, 5))
plt.xlabel('x')
plt.ylabel('index')
plt.scatter(x, index)

avg = np.mean(x)
md = np.median(x)
print('Среднее значение для массива =', avg)
print('Среднее медианное значение для массива =', md)
if avg > md:
  print("Среднее арифметическое больше")
elif avg < md:
  print("Среднее медианное больше")
else:
  print("Значения равны")
plt.show()
```

## Задание 4.3.2

Дана функция $\frac{\sqrt{1+e^{\sqrt{x}} +\cos x^{2}}}{|1-\sin^{3} |} +\ \ln |2x|$;
Создать массив из 10 значений функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой половины массива и построить графики
для основного массива – линейный и для среза – точечный

```python
N = 10
arr = []
w = []
for x in range(1, N + 1):
    arr.append(np.sqrt(1 + np.e ** np.sqrt(x) + np.cos(x ** 2)) / abs(1 - np.sin(x) ** 3))
print(arr)
slice = arr[:len(arr) // 2]
for i in range(1, len(slice) + 1):
    w.append(i)
print(w)
print(slice)

plt.plot(arr)
plt.xlabel("x")
plt.ylabel("y")
plt.show()

plt.scatter(w, slice)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
